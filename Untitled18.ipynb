 # Ejemplo 1
 * El Contador: Este ejemplo muestra cómo repetir una acción un número específico de veces basándose en un límite que tú decidas.

```python
print("--- Ejemplo 1: Contador Personalizado ---")
limite = int(input("¿Hasta qué número quieres contar? "))
contador = 1

while contador <= limite:
    print("Contando:", contador)
    contador += 1  # Incrementamos para no crear un ciclo infinito
print("Finalizando conteo del 1 al",limite)
print("Programa realizado por David Barrera 1300")
```

# Ejemplo 2
* El Centinela (Validación de Datos)
Aquí el ciclo no se detiene hasta que el usuario introduce una "palabra clave" o un número específico. Es ideal para menús o salidas de programas.

```python
print("\n--- Ejemplo 2: El Centinela ---")
numero = 0

while numero != 7:
    numero = int(input("Adivina el número de la suerte para salir, Pista: es un numero del uno a 10: "))
    if numero != 7:
        print("Casi, pero no. ¡Intenta de nuevo!")

print("¡Acertaste! Saliendo del ciclo...")
print("Programa realizado por David Barrera 1300")
```

# Ejemplo 3
* Acumulador de Suma
En este caso, usamos el ciclo para ir sumando valores a una bolsa total hasta que decidamos parar.

```python
print("\n--- Ejemplo 3: Acumulador de Suma ---")
total_suma = 0
cantidad_a_sumar = int(input("¿Cuántos números quieres sumar en total? "))
vueltas = 0

while vueltas < cantidad_a_sumar:
    valor = int(input("Introduce un número para sumar: "))
    total_suma += valor
    vueltas += 1

print("La suma total de tus números es:", total_suma)
print("Programa realizado por David Barrera 1300")
```

# Ejemplo 4
* Control de "Presupuesto" (Resta sucesiva)
Este ejemplo simula una compra. El ciclo sigue funcionando mientras todavía tengas "dinero" disponible.

```python
print("\n--- Ejemplo 4: Control de Presupuesto ---")
presupuesto = int(input("¿Cuánto dinero tienes en tu cartera? "))

while presupuesto > 0:
    gasto = int(input(f"Tienes {presupuesto} disponible. ¿Cuánto cuesta este producto? "))

    if gasto > presupuesto:
        print("No tienes suficiente dinero para eso.")
    else:
        presupuesto -= gasto
        print("Compra realizada.")
        break

print("Te has quedado sin presupuesto.")
print("Programa realizado por David Barrera 1300")
```

# Ejemplo 5
* Generador de Tablas de Multiplicar
Usamos el ciclo para generar una tabla matemática completa de forma dinámica.

```python
print("\n--- Ejemplo 5: Tabla de Multiplicar ---")
tabla_del = int(input("¿De qué número quieres ver la tabla de multiplicar? "))
multiplicador = 1

while multiplicador <= 10:
    resultado = tabla_del * multiplicador
    print(f"{tabla_del} x {multiplicador} = {resultado}")
    multiplicador += 1
print("fin de la tabla")
print("Programa realizado por David Barrera 1300")
```
